```python
import math
import tkinter as tk
from functools import *

def input():
    I2.set(I.get())
    I1.set(myText.get())
    res.set(result())
    
def result():          # using if else statments to use the correct operator on inputs
    x = I1.get()
    y = I2.get()
    if op.get() == 1:                      # 1 for addition
        res=int(I1.get())+int(I2.get())   
        myText.set(res)
        
    elif op.get() == 2 :                   # 2 for subtraction
        res=int(I1.get())-int(I2.get())
        myText.set(res)
     
    elif op.get() == 3 :                   # 3 for multiplication
        res=int(I1.get())*int(I2.get())
        myText.set(res)
    
    elif op.get() == 4 :                   # 4 for division
        res=int(I1.get())//int(I2.get())
        myText.set(res)
    
    elif op.get() == 5 :                   # 5 for hcf
        if x > y:
            smaller = y
        else:
            smaller = x
        for i in range(1, smaller+1):
            if((x % i == 0) and (y % i == 0)):
                myText.set(i)
                
    elif op.get() == 6 :                   # 6 for lcm
        def compute_gcd(x, y):
            while(y):
                 x, y = y, x % y
            return x
        lcm = (x*y)//compute_gcd(x,y)    
        myText.set(lcm)
     
    elif op.get() == 7 :                   # 7 for power
        myText.set(x**y)
        
    elif op.get() == 8 :                   # 8 for log
        myText.set(math.log(x,y))    
    
    else:
        myText.set("Invalid operation")

root=tk.Tk()
root.title('CALCULATOR')
root.geometry("700x600")   # the size of console window will be 400x400

# Labels for info of operations
tk.Label(root,text='CALCULATOR',padx = 10,bg="white").grid(row=1, column=21)
tk.Label(root,text='Please select operations -> ',padx = 10,bg="white").grid(row=2, column=20)
tk.Label(root,text='1 for Addition  ,  5 for hcf',padx = 10,bg="white").grid(row=3, column=20)
tk.Label(root,text='2 for subtraction  ,  6 for lcm',padx = 10,bg="white").grid(row=4, column=20)
tk.Label(root,text='3 for multiplication  ,  7 for power',padx = 10,bg="white").grid(row=5, column=20)
tk.Label(root,text='4 for division  ,  8 for Logarithm (base x of y)',padx = 10,bg="white").grid(row=6, column=20)

# taking the important variables including inputs
res = tk.IntVar()
I = tk.IntVar()
I1 = tk.IntVar()    # 1st input
I2 = tk.IntVar()    # 2nd input
op = tk.IntVar()    # operation input
myText=tk.StringVar()  # for output 


tk.Label(root, text="Enter 1st Number :",padx = 10).grid(row=8, column=22)
tk.Label(root, text="Enter 2nd Number :",padx = 10).grid(row=9, column=22)
tk.Label(root, text="Enter Operation :",padx = 10).grid(row=10, column=22)

tk.Entry(root, textvariable=I1,justify="center",fg="#39ff14").grid(row=8, column=23)
tk.Entry(root, textvariable=I2,justify="center",fg="#39ff14").grid(row=9, column=23)
tk.Entry(root, textvariable=op,justify="center",fg="#39ff14").grid(row=10, column=23)

button = tk.Button(root, text="Calculate",command=result)
button.grid(row=12, column=23)
tk.Label(root,text='Result : ',bg="white").grid(row=15, column=22)
tk.Label(root, text="", textvariable=myText,fg="red",bg="black").grid(row=15, column=23)  # to get the result as output

tk.Label(root, text=" new input (click on YES) : ",padx = 10).grid(row=17, column=20)
tk.Entry(root, textvariable=I,justify="center",fg="#39ff14").grid(row=17, column=21)
button = tk.Button(root, text="YES",command=input)
button.grid(row=18, column=21)
tk.Label(root,text='Click on YES to get new Result : ').grid(row=18, column=20)
tk.Label(root,text='new Result : ',bg="white").grid(row=19, column=20)
tk.Label(root, text="", textvariable=myText,fg="red",bg="black").grid(row=19, column=21)    

root.mainloop()
```

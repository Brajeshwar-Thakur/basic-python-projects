```python
import json
from functools import *
from tkinter import *
from tkinter import messagebox as mb
from PIL import ImageTk,Image
```

# <center><font face="courier"> <font color=#34FFB6 > Logic </font> for Quiz </font><center>

```python
class Quiz:     
    
    # this works as the constructor , helps in initializing the values for variables
    def __init__(self):
        
        # this helps in showing the right answer
        self.index = 0 # initializing index to 0
         
        self.que_no = 0 # initializing question number to 0
         
        # assigns ques to the Show_Que function to update later.
#       Label(root, text="QUIZ !!",padx = 10,width=50, bg="grey",fg="#39ff14", font=("courier", 24, "bold")
#       ).grid(row=8, column=20)
        self.Show_Que()
         
        # chosen_opt holds an integer value which is used for      
        self.chosen_opt = IntVar() # this is variable for the selected option in a question.
         
        # displaying radio button for the current question   
        self.opts = self.radio_buttons() # displays options for the current question
         
        # display options for the current question
        self.Show_options()
         
        # displays the button for next and exit.
        self.buttons()
         
        # no of questions
        self.data_size = len(question)
         
        # keep a counter of Ans_counter answers
        self.Ans_counter = 0
    
    # This method is used to display the right ans after selecting the option and clicking next
    def message_box_Ans(self): 
        ans = ["Chandigarh","Delhi","Goa","Agra"]      # list cointaining all the answers
        mb.showinfo("ans : ",ans[self.index])          # creates a message box to show the correct ans
        # helps in looping through the answers
        self.index += 1
            
            
 
    # This method is used to display the end result as in the no. of correct attempts 
    def End_result(self):
         
        # calculates the wrong count
        wrong_count = self.data_size - self.Ans_counter
        Ans_counter = f"Ans_counter: {self.Ans_counter}"
        wrong = f"Wrong: {wrong_count}"
         
        # calcultaes the percentage of Ans_counter answers
        score = int(self.Ans_counter / self.data_size * 100)
        result = f"Score: {score}%"
         
        # Shows a message box to display the result
        mb.showinfo("Result", f"{result}\n{Ans_counter}\n{wrong}")
 
 
    # This method checks the Answer after we click on Next.
    def check_ans(self, que_no):
         
        # checks for if the selected option is Ans_counter
        if self.chosen_opt.get() == answer[que_no]:
            # if the option is Ans_counter it return true
            print("Ans_counter ans is : ",answer[que_no])
            return True
 
    # This method is used to move to the next que and check the answer of the and increment counter
    def next_btn(self):
         
        # Check if the answer is Ans_counter
        if self.check_ans(self.que_no):
             
            # if the answer is Ans_counter it increments the Ans_counter by 1
            self.Ans_counter += 1
         
        # Moves to next Question by incrementing the que_no counter
        self.que_no += 1
         
        # checks if the que_no size is equal to the data size
        if self.que_no==self.data_size:
             
            # if it is Ans_counter then it displays the score
            self.End_result()
             
            # destroys the root
            root.destroy()
        else:
            # shows the next question
            self.Show_Que()
            self.Show_options()  
        self.message_box_Ans()
#         ans = ["Chandigarh","Delhi","Goa","Agra"]
#         mb.showinfo("ans : ",ans[self.index])
#         self.index += 1
 
 
    # This method is used to display the two buttons on the screen i.e., quit and next
    def buttons(self):
         
        # The first button is the Next button to move to the
        # next Question
        next_button = Button(root, text="Next Que",command=self.next_btn,
        width=15,height=2,bg="#34FFB6",fg="white",font=("courier",16," bold"))
         
        # palcing the button  on the screen
        next_button.place(x=350,y=380)
         
        # This is the second button which is used to Quit the root
        quit_button = Button(root, text="Quit game", command=root.destroy,
        width=10,bg="black", fg="red",font=("courier",16," bold"))
         
        # placing the Quit button on the screen
        quit_button.place(x=700,y=50)
        
#         mb.showinfo("ans : ",answer)
 
 
    # This method deselects the options(radio button) on the screen 
    def Show_options(self):
        val=0
         
        # deselecting the options
        self.chosen_opt.set(0)
         
        # looping over the options to be displayed for the
        # text of the radio buttons.
        for option in options[self.que_no]:
            self.opts[val]['text']=option
            val+=1
 
 
    # This method shows the current Question on the screen
    def Show_Que(self):
         
        # setting the Question properties
        que_no = Label(root, text=question[self.que_no], width=60,
        font=( 'courier' ,16, 'bold' ), anchor= 'w' )
         
        #placing the option on the screen
        que_no.place(x=70, y=100)
 
 
    # This method shows the radio buttons to select the Question
    def radio_buttons(self):
         
        # initialize the list with an empty list of options
        q_list = []
         
        # position of the first option
        y_pos = 150
         
        # adding the options to the list
        while len(q_list) < 4:
             
            # setting the radio button properties
            radio_btn = Radiobutton(root,text=" ",variable=self.chosen_opt,
            value = len(q_list)+1,font = ("courier",14))
             
            # adding the button to the list
            q_list.append(radio_btn)
             
            # placing the button
            radio_btn.place(x = 100, y = y_pos)
             
            # incrementing the y-axis position by 40
            y_pos += 40
         
        # return the radio buttons
        return q_list
        
```


# <font face="courier"><center> <font color=#39ff14 > GUI  </font> Window </font>

```python
root=Tk()                     # starting/creating the quiz window
root.title('Quiz')            # name is displayed at the title bar
root.geometry("950x600")      # the size of console window 

# lable is used to display the title quiz 
Label(root, text="QUIZ !!",padx = 10,width=50, bg="grey",fg="#39ff14", font=("courier", 24, "bold")).grid(row=8, column=20)

# C = Canvas(root, bg="blue", height=250, width=300)
# filename = PhotoImage(file = "quiz.jpg")
# background_label = Label(root, image=filename)
# background_label.place(x=0, y=0, relwidth=1, relheight=1)

with open('data.json') as f:  # utilizing the json file data
    data = json.load(f)
 
# set the question, options, and answer
question = (data['question'])
options = (data['options'])
answer = (data[ 'answer'])

obj = Quiz()                  # creating an object 
root.mainloop()               # ending the window
```
